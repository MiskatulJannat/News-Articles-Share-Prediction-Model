```python
import numpy as np
import pandas as pd
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)

import matplotlib.pyplot as plt
import seaborn as sns

import os
import time 
import re

import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn import metrics

from scipy import stats

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('dataset/news_article_ready.csv')
print(df.shape)
df.head()
```

```python
X = df.drop('shares',axis=1)
y = df[['shares']]

print(X.shape)
print(y.shape)

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=163)
print((X_train.shape, X_test.shape), (y_train.shape,y_test.shape))
```

# Model Training

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Model training

# Create a Linear Regression model
linear_reg = LinearRegression()

# Fit the model to the training data
linear_reg.fit(X_train, y_train)

y_train_pred = linear_reg.predict(X_train)

# Evaluate the trained model
train_rmse = np.sqrt(mean_squared_error(y_train, y_train_pred))
print("Linear Training RMSE:", train_rmse)

train_r2 = r2_score(y_train, y_train_pred)
print("Linear Training R^2 Score:", train_r2)
```

```python
# Make predictions on the testing data
y_test_pred = linear_reg.predict(X_test)

# Evaluate the model
test_rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))
print("Linear Testing RMSE:", test_rmse)

test_r2 = r2_score(y_test, y_test_pred)
print("Linear Testing R^2 Score:", test_r2)
```

#### Note: RMSE is very high, means this model's predictions are far from the actual values. R^2 also far from 1 and 0.147 is relatively low, indicating that there is a lot of unexplained variance in the data.
