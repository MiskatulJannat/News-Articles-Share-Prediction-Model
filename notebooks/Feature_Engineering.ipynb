### Adding noun_count, verb_count, and adjective_count features

```python
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```python
import nltk
from nltk import word_tokenize, pos_tag
```

```python
# Function to count nouns in a sentence
def count_nouns(text):
    # Tokenize the text and get part-of-speech tags
    tokens = word_tokenize(text)
    pos_tags = pos_tag(tokens)
    
    # Count nouns (tags starting with 'N')
    noun_count = sum(1 for _, tag in pos_tags if tag.startswith('N'))
    return noun_count
```

```python
# Apply the function to each row in the 'text + text' column and create the 'noun_count' column
df1['noun_count'] = df1['title + text'].apply(count_nouns)

df1.sample(6)
```

```python
# Function to count verbs in a sentence
def count_verbs(text):
    # Tokenize the text and get part-of-speech tags
    tokens = word_tokenize(text)
    pos_tags = pos_tag(tokens)
    
    # Count verbs (tags starting with 'V')
    verb_count = sum(1 for _, tag in pos_tags if tag.startswith('V'))
    return verb_count

# Apply the function to each row in the 'title + text' column and create the 'verb_count' column
df1['verb_count'] = df1['title + text'].apply(count_verbs)

df1.sample(3)
```

```python
# Function to count adjectives in a sentence
def count_adjectives(text):
    # Tokenize the text and get part-of-speech tags
    tokens = word_tokenize(text)
    pos_tags = pos_tag(tokens)
    
    # Count adjectives (tags starting with 'JJ')
    adjective_count = sum(1 for _, tag in pos_tags if tag.startswith('JJ'))
    return adjective_count

# Apply the function to each row in the 'title + text' column and create the 'adjective_count' column
df1['adjective_count'] = df1['title + text'].apply(count_adjectives)

df1.tail()
```

### Using NGRAM to generate new features

```python
combined_text = ' '.join(df1['title + text'])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(ngram_range=(2,2),max_features=10)
```

```python
# Fit the CountVectorizer with the 'combined_text' data and transform 'df1' text data
X_combined = cv.fit_transform([combined_text])
X_df1 = cv.transform(df1['title + text'])
```

```python
# Create a DataFrame 'df_ngram' to store the bag of words for each text in 'df1'
df_ngram = pd.DataFrame(X_df1.toarray(), columns=cv.get_feature_names_out())
```

```python
# Adding this to the main dataframe
df_final = pd.concat([df1.reset_index(drop=True),df_ngram.reset_index(drop=True)],axis=1)
df_final.shape
```
