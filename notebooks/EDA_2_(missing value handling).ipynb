```python
# Preprocessing the text data
REPLACE_BY_SPACE_RE = re.compile("[/(){}\[\]\|@,;!]")
BAD_SYMBOLS_RE = re.compile("[^0-9a-z #+_]")
STOPWORDS_nlp = set(stopwords.words('english'))


# Customizing Stopword list
stoplist = ["i", "project", "living", "home", 'apartment', "pune", "me", "my", "myself", "we", "our", "ours",
            "ourselves", "you", "you're", "you've", "you'll", "you'd", "your", "yours", "yourself", "yourselves",
            "he", "him", "his", "himself", "she", "she's", "her", "hers", "herself", "it", "it's", "its", "itself",
            "they", "them", "their", "theirs", "themselves", "what", "which", "who", "whom", "this", "that",
            "that'll", "these", "those", "am", "is", "are", "was", "were", "be", "been", "being", "have", "has",
            "had", "having", "do", "does", "did", "doing", "a", "an", "the", "and", "but", "if", "or", "because",
            "as", "until", "while", "of", "at", "by", "for", "with", "about", "against", "between", "into", "through",
            "during", "before", "after", "above", "below", "to", "from", "up", "down", "in", "out", "on", "off", "over",
            "under", "again", "further", "then", "once", "here", "there", "when", "where", "why", "all", "any", "both",
            "each", "few", "more", "most", "other", "some", "such", "no", "nor", "not", "only", "own", "same", "so",
            "than", "too", "very", "s", "t", "can", "will", "just", "don", "don't", "should", "should've", "now", "d",
            "ll", "m", "o", "re", "ve", "y", "ain", "aren", "couldn", "didn", "doesn", "hadn", "hasn", "haven", "isn",
            "ma", "mightn", "mustn", "needn", "shan", "shan't", "shouldn", "wasn", "weren", "won", "rt", "rt", "qt", "for",
            "the", "with", "in", "of", "and", "its", "it", "this", "i", "have", "has", "would", "could", "you", "a", "an",
            "be", "am", "can", "edushopper", "will", "to", "on", "is", "by", "ive", "im", "your", "we", "are", "at", "as",
            "any", "ebay", "thank", "hello", "know", "need", "want", "look", "hi", "sorry", "http", "https", "body",
            "dear", "hello", "hi", "thanks", "sir", "tomorrow", "sent", "send", "see", "there", "welcome", "what", "well", "us"]

# Load NLTK stopwords
STOPWORDS_nlp = set(stopwords.words('english'))

# Custom preprocessing function
def text_prepare(text):
    """
    text: a string
    return: modified initial string
    """
    text = re.sub(r"\d+", " ", text)  # removing digits
    text = re.sub(r"(?:\@|https?\://)\S+", "", text)  # removing mentions and URLs
    text = text.lower()  # lowercase text
    text = re.sub('[0-9]+', '', text)
    text = re.sub(REPLACE_BY_SPACE_RE, " ", text)  # replace REPLACE_BY_SPACE_RE symbols by space in text
    text = re.sub(BAD_SYMBOLS_RE, " ", text)  # delete symbols which are in BAD_SYMBOLS_RE from text
    text = ' '.join([word for word in text.split() if word not in STOPWORDS_nlp])  # delete NLTK stopwords from text
    text = ' '.join([word for word in text.split() if word not in stoplist])  # delete custom stopwords from text
    text = text.strip()
    return text

# Apply the preprocessing function to "title" and "text" columns and create new columns
df1['title_cleaned'] = df1['title'].astype(str).apply(text_prepare)
df1['text_cleaned'] = df1['text'].astype(str).apply(text_prepare)

# Check the updated DataFrame
print(df1[['title_cleaned', 'text_cleaned']])
```

```python
# Adding a new column combining cleaned title and text column
df1['title + text'] = df1['title_cleaned'] + ' ' + df1['text_cleaned']

df1.shape
```

### Since we have so many missing values, we can fill them using the heavily used words in 'title' and 'text' columns in each classes (business, tech....etc).

```python
# Filter rows where 'data_channel' is "business" and concatenate the text
temp_business_text = ' '.join(df1[df1['data_channel'] == 'business']['title + text'].tolist())

# Check the merged text
print(temp_business_text)
```

```python
# Define a function to extract top 50% word used in each data_channel class

from collections import Counter

def get_top_words_by_data_channel(data_frame, data_channel):
    # Filter rows where 'data_channel' matches the input data_channel and concatenate the text
    channel_text = ' '.join(data_frame[data_frame['data_channel'] == data_channel]['title + text'].tolist())

    # Tokenize the text
    channel_text_words = channel_text.split()

    # Count word frequency
    word_freq = Counter(channel_text_words)

    # Calculate the number of unique words
    total_unique_words = len(word_freq)

    # Calculate the number of words to keep (top 50%)
    top_percentage = 0.5
    num_words_to_keep = int(total_unique_words * top_percentage)

    # Select the top 30% of words by frequency
    top_words = [word for word, freq in word_freq.most_common(num_words_to_keep)]

    # Join the top words back into a string
    top_words_text = ' '.join(top_words)

    return top_words_text
```

```python
# Top words in business
business_top_words = get_top_words_by_data_channel(df1, 'business')
print("There are {} words in the text.".format(len(business_top_words)))
```

```python
# Top words in entertainment
entertainment_top_words = get_top_words_by_data_channel(df1, 'entertainment')
print("There are {} words in the text.".format(len(entertainment_top_words)))
```

```python
# Top words in tech
tech_top_words = get_top_words_by_data_channel(df1, 'tech')
print("There are {} words in the text.".format(len(tech_top_words)))
```

```python
# Top words in lifestyle
lifestyle_top_words = get_top_words_by_data_channel(df1, 'lifestyle')
print("There are {} words in the text.".format(len(lifestyle_top_words)))
```

### Cosine similarity checking

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df3 = df1.copy()    # For testing purpous
```

```python
# Check which class has most cosine similarity

text_column_name = 'title + text'

# Extract the text in the row where 'data_channel' is '00000'
text_00000 = df3.loc[df3['data_channel'] == '00000', text_column_name].values[0]


# Create a TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fit and transform the top words and '00000' text
tfidf_matrix = tfidf_vectorizer.fit_transform([business_top_words, entertainment_top_words, tech_top_words,
                                               social_media_top_words, lifestyle_top_words, text_00000])

# Calculate the cosine similarity between '00000' text and top words for each category
cosine_similarities = cosine_similarity(tfidf_matrix[5], tfidf_matrix[:-1])

# Find the category with the highest cosine similarity
most_similar_category = ["business", "entertainment", "tech", "social media", "lifestyle"][cosine_similarities.argmax()]


print(most_similar_category)
cosine_similarities
```

### Fill the missing values in the 'data_channel' column with the class that has the highest cosine similarity.

```python
# Create a TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Create a function to calculate the most similar category
def get_most_similar_category(text):
    tfidf_matrix = tfidf_vectorizer.fit_transform([business_top_words, entertainment_top_words, tech_top_words,
                                                   social_media_top_words, lifestyle_top_words, text])
    
    cosine_similarities = cosine_similarity(tfidf_matrix[5], tfidf_matrix[:-1])
    most_similar_category_idx = cosine_similarities.argmax()
    
    categories = ["business", "entertainment", "tech", "social media", "lifestyle"]
    
    return categories[most_similar_category_idx]

# Replace '00000' values in the 'data_channel' column with the most similar category
for index, row in df1.iterrows():
    if row['data_channel'] == '00000':
        text_00000 = row['title + text']
        most_similar_category = get_most_similar_category(text_00000)
        df1.at[index, 'data_channel'] = most_similar_category

df1.head()
```
